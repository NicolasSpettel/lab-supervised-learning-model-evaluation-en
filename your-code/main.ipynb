## Regression Model Evaluation

```python
column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
data = pd.read_csv('housing.csv', header=None, delimiter=r"\s+", names=column_names)
```

```python
"""
CRIM - per capita crime rate by town
ZN - proportion of residential land zoned for lots over 25,000 sq.ft.
INDUS - proportion of non-retail business acres per town.
CHAS - Charles River dummy variable (1 if tract bounds river; 0 otherwise)
NOX - nitric oxides concentration (parts per 10 million)
RM - average number of rooms per dwelling
AGE - proportion of owner-occupied units built prior to 1940
DIS - weighted distances to five Boston employment centres
RAD - index of accessibility to radial highways
TAX - full-value property-tax rate per $10,000
PTRATIO - pupil-teacher ratio by town
B - 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town
LSTAT - % lower status of the population
MEDV - Median value of owner-occupied homes in $1000's"""
```

```python
data
```

## 1. Split this data set into training (80%) and testing (20%) sets.

The `MEDV` field represents the median value of owner-occupied homes (in $1000's) and is the target variable that we will want to predict.

```python
X = data.drop(columns = "MEDV")
y = data["MEDV"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

## 2. Train a `LinearRegression` model on this data set and generate predictions on both the training and the testing set.

```python
from sklearn import linear_model
lin = linear_model.LinearRegression()
lin.fit(X_train, y_train)
```

```python
y_train_pred = lin.predict(X_train)
```

```python
y_test_pred = lin.predict(X_test)
```

## 3. Calculate and print R-squared for both the training and the testing set.

```python
from sklearn.metrics import r2_score
r2_train = r2_score(y_train, y_train_pred)
r2_test = r2_score(y_test, y_test_pred)
print(f"r2-score for train data: {r2_train}")
print(f"r2-score for tesat data: {r2_test}")
```

## 4. Calculate and print mean squared error for both the training and the testing set.

```python
from sklearn.metrics import mean_squared_error
mse_train = mean_squared_error(y_train, y_train_pred)
mse_test = mean_squared_error(y_test, y_test_pred)
print(f"mse for train data: {mse_train}")
print(f"mse for tesat data: {mse_test}")
```

## 5. Calculate and print mean absolute error for both the training and the testing set.

```python
from sklearn.metrics import mean_absolute_error
mae_train = mean_squared_error(y_train, y_train_pred)
mae_test = mean_squared_error(y_test, y_test_pred)
print(f"mae for train data: {mae_train}")
print(f"mae for tesat data: {mae_test}")
```

## Classification Model Evaluation

```python
from sklearn.datasets import load_iris
data = load_iris()
```

```python
print(data.DESCR)
```

```python
column_names = data.feature_names
```

```python
df = pd.DataFrame(data['data'],columns=column_names)
```

```python
df
```

```python
target = pd.DataFrame(data.target)
```

```python
data.keys()
```

```python
data['target_names']
```

## 6. Split this data set into training (80%) and testing (20%) sets.

The `class` field represents the type of flower and is the target variable that we will want to predict.

```python
data
```

```python
X = df
y = target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

## 7. Train a `LogisticRegression` model on this data set and generate predictions on both the training and the testing set.

```python
from sklearn import linear_model
log = linear_model.LogisticRegression()
log.fit(X_train, y_train)
y_train_pred = log.predict(X_train)
y_test_pred = log.predict(X_test)
```

## 8. Calculate and print the accuracy score for both the training and the testing set.

```python
from sklearn.metrics import accuracy_score, balanced_accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
print(accuracy_score(y_train, y_train_pred))
print(accuracy_score(y_test, y_test_pred))
```

## 9. Calculate and print the balanced accuracy score for both the training and the testing set.

```python
print(balanced_accuracy_score(y_train, y_train_pred))
print(balanced_accuracy_score(y_test, y_test_pred))
```

## 10. Calculate and print the precision score for both the training and the testing set.

```python
print(precision_score(y_train, y_train_pred, average = "weighted"))
print(precision_score(y_test, y_test_pred, average = "weighted"))
```

## 11. Calculate and print the recall score for both the training and the testing set.

```python
print(recall_score(y_train, y_train_pred, average = "weighted"))
print(recall_score(y_test, y_test_pred, average = "weighted"))
```

## 12. Calculate and print the F1 score for both the training and the testing set.

```python
print(f1_score(y_train, y_train_pred, average = "weighted"))
print(f1_score(y_test, y_test_pred, average = "weighted"))
```

## 13. Generate confusion matrices for both the training and the testing set.

```python
print(confusion_matrix(y_train, y_train_pred))
print(confusion_matrix(y_test, y_test_pred))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(confusion_matrix(y_train, y_train_pred), xticklabels = ['setosa', 'versicolor', 'virginica'], yticklabels = ['setosa', 'versicolor', 'virginica'], annot=True)
plt.show()
sns.heatmap(confusion_matrix(y_test, y_test_pred), xticklabels = ['setosa', 'versicolor', 'virginica'], yticklabels = ['setosa', 'versicolor', 'virginica'],  annot=True)
```

```python
print(classification_report(y_train, y_train_pred))
print(classification_report(y_test, y_test_pred))
# to see all at once
```
